sklearn을 활용한 다중 회귀 분석을 토대로 한 자동차 데이터 분석
 - 차의 무게와 부피와 이산화탄소 배출량의 관계를 알아보자.

## 1. 라이브러리

```python
# 코드 작성: pandas 라이브러리 pd로 import하기
import pandas as pd
```

## 2. 데이터 준비하기

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
# 코드 작성: car라는 변수에 cars.csv 읽어오기(io 활용)
car = pd.read_csv(io.BytesIO(uploaded['cars.csv']))
```

```python
# 코드 작성: car 출력
car
```

```python
# 코드 작성: 변수 X는 car의 무게(Weight)와 부피(Volume) / Hint: DataFrame 형태로 추출하기!
X = car[['Weight', 'Volume']]
# 코드 작성: 변수 Y는 car의 이산화탄소 배출량(CO2) / Hint: DataFrame 형태로 추출하기!
Y = car[['CO2']]
```

## 3. 회귀 분석


*   **sklearn**: 회귀 분석을 돕는 라이브러리
*   **linear_model**: sklearn 내 선형 모형을 토대로 선형 회귀 분석을 도움,


```python
from sklearn import linear_model
```

*   **LinearRegression()**: 선형 회귀 분석 메소드
*   **fit(X, Y)**: 모델에 데이터 X, Y를 맞추는 함수 

```python
regr = linear_model.LinearRegression()
# 코드 작성: regr을 X, Y 데이터에 맞춰 fit 시키기
regr.fit(X, Y)
```

*   **predict(X, Y)**: 

```python
# 코드 작성: 무게가 2300kg(X), 부피가 1300cm^3(Y)인 차의 이산화탄소 배출량(Y) 예측해보기
predictedCO2 = regr.predict([[2300, 1300]])
print(predictedCO2)
```



*   **.coef_**: 기울기, 즉 계수



```python
# 코드 작성: regr의 계수 출력해보기
print(regr.coef_)
```
