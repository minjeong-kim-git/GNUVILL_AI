## 1. 데이터 만들기

```
# 라이브러리 
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
```

```
from sklearn import datasets
boston_house_price = datasets.load_boston()
```

```
print(boston_house_price)
```

```
# 데이터 파악
print(boston_house_price.keys())
print(boston_house_price.data.shape)
print(boston_house_price.feature_names)
```

컬럼 설명

컬럼.png

```
print(boston_house_price.DESCR)
```

## 2. 데이터프레임 만들기

```
df = pd.DataFrame(boston_house_price.data)
df.head()
```

```
df.columns = boston_house_price.feature_names
df.head()
```

```
df['Price'] = boston_house_price.target
df.head()
```

##3. 방의 수와 집값 관계 알아보기

```
df.plot(kind = 'scatter', x='RM', y = 'Price', figsize = (7,7), color = 'red')
```

```
# 그래프 자르기
df.plot(kind = 'scatter', x='RM', y = 'Price', figsize = (7,7), color = 'red',
        xlim = (4, 8), ylim = (10, 45))
```

##4. 단순선형회귀식

```
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()

# 풀이 1
linear_regression.fit(X = pd.DataFrame(df['RM']), y = df['Price'])
prediction = linear_regression.predict(X = pd.DataFrame(df['RM']))

# 변수 alpha에 절편 저장 후 출력
print('alpha = ', linear_regression.intercept_)
# 변수 beta에 기울기 저장 후 출력
print('beta = ', linear_regression.coef_)
```

```
# 풀이 2

from sklearn import linear_model
linear_regression = linear_model.LinearRegression()
linear_regression.fit(X = pd.DataFrame(df['RM']), y = df['Price'])
prediction = linear_regression.predict(X = pd.DataFrame(df['RM']))

# 변수 alpha에 절편 저장 후 출력
print('alpha = ', linear_regression.intercept_)
# 변수 beta에 기울기 저장 후 출력
print('beta = ', linear_regression.coef_)
```

##5. 산점도에 선형회귀곡선도를 그려 관계 파악

```
# 코드 작성

# 산포도
df.plot(kind = 'scatter', x='RM', y = 'Price', figsize = (7,7), color = 'red',
        xlim = (4, 8), ylim = (10, 45))

# 선형회귀곡선도
plt.plot(df['RM'], prediction, color='blue')
```
